```python
def BS_fix(bitsize):
    standard_bs_vals = [26, 17.5, 17, 14.75, 12.5, 12.25, 9.875, 8.5, 8.375, 6.5, 6]
    if bitsize >= 25 and bitsize <= 27:
        return 26
    elif bitsize >= 17.3 and bitsize <= 24:
        return 17.5
    elif bitsize >= 16.0 and bitsize < 17.3:
        return 17
    elif bitsize >= 14 and bitsize <= 15:
        return 14.75
    elif bitsize >= 12.3 and bitsize <= 13:
        return 12.5
    elif bitsize >= 11.8 and bitsize < 12.3:
        return 12.25
    elif bitsize >= 9.4 and bitsize <= 11.5:
        return 9.875
    elif bitsize >= 8.3 and bitsize < 9.4:
        return 8.5
    elif bitsize >= 8 and bitsize < 8.3:
        return 8.25
    elif bitsize >= 6.3 and bitsize <=7:
        return 6.75
    elif bitsize >= 5.8 and bitsize < 6.3:
        return 6
    else:
        return np.nan
```

```python
CALI_DF['BS_FIX'] = CALI_DF.apply(lambda x: BS_fix(x.loc['BS']), axis=1)
```

```python
# Using new function increased BS_FIX from around 387000 to 682621
CALI_DF.info()
```

Attempting to create a rolling mean column for caliper. Needs to be a large enough window to account for large washout sections.For this we use .rolling   
Once calculated, it can be passed into the BS_fix function and checked witha crossplot

```python
CALI_DF['CALI_R_MEAN'] = CALI_DF['CALI'].rolling(5).mean()
```

```python
CALI_DF['BS_FROM_CALI'] = CALI_DF.apply(lambda x: BS_fix(x.loc['CALI']), axis=1)
```

```python
CALI_DF['BS_COMB']=CALI_DF['BS_FIX']
```

```python
CALI_DF['BS_COMB'].fillna(CALI_DF['BS_FROM_CALI'], inplace=True)
```

```python
CALI_DF['DIFF_CAL']=CALI_DF['CALI'] - CALI_DF['BS_COMB']
```

```python
CALI_DF['DIFF_CAL'].max()
```

```python
CALI_DF.head(5000)
```

```python
CALI_DF['BADHOLE_CAL'] = np.where(CALI_DF['DIFF_CAL']>=1.5, 1, 0)
```

```python
CALI_DF['BADHOLE_CAL'].value_counts()
```

```python
CALI_DF.plot(kind='scatter', x='BS', y='BS_FROM_CALI')
```

```python
mean_cali_by_group = CALI_DF.groupby('GROUP')['CALI'].mean()
```

```python
CALI_DF['MEAN_CALI'] = CALI_DF['GROUP'].map(mean_cali_by_group)
```

```python
CALI_DF.describe()
```

```python
def synth_BS(mean_caliper):
    if mean_caliper >=16 and mean_caliper <=23:
        bitsize = 17.5
    elif mean_caliper >12 and mean_caliper < 16:
        bitsize = 12.25
    elif mean_caliper >= 8 and mean_caliper < 12.25:
        bitsize = 8.5
    elif mean_caliper >= 5.5 and mean_caliper < 8.5:
        bitsize = 6
    else:
        return np.nan
        
    return bitsize
```

```python
CALI_DF['SYNTH_BS'] = CALI_DF.apply(lambda x: synth_BS(x['MEAN_CALI']), axis=1)
```
